```python
import psycopg2

conn = psycopg2.connect(dbname="anyway", user="anyway", password="anyway", host="db")
cur = conn.cursor()
query = "SELECT * FROM markers_hebrew LIMIT 100"
cur.execute(query)
markers = cur.fetchall()
```

```python
column_names = [desc[0] for desc in cur.description]
for column in column_names:
    print(column)
```

```python
import pandas as pd
df_markers_hebrew = pd.DataFrame(markers, columns = column_names)
df_markers_hebrew.set_index('id', inplace=True)
```

```python
query = "SELECT * FROM markers_hebrew WHERE (accident_day = 2 AND accident_month = 9 AND accident_year = 2019)"
cur.execute(query)
accident_markers_hebrew_from_02_09_2019 = cur.fetchall()
```

```python
df_from_02_09_2019 = pd.DataFrame(accident_markers_hebrew_from_02_09_2019, columns = column_names)
df_from_02_09_2019.set_index('id', inplace=True)
```
