```python
import psycopg2

conn = psycopg2.connect(dbname="anyway", user="anyway", password="anyway", host="db")
cur = conn.cursor()
```

```python
query = "SELECT * FROM markers ORDER BY created ASC LIMIT 1000"
cur.execute(query)
markers = cur.fetchall()
```

```python
column_names = [desc[0] for desc in cur.description]
for column in column_names:
    print(column)
```

```python
import pandas as pd
df_markers = pd.DataFrame(markers, columns = column_names)
df_markers.set_index('id', inplace=True)
```
